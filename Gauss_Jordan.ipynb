```python
import numpy as np

a = [[8, 3, -3], [-2, -8, 5], [3, 5, 10]]

# Untuk mencari koefisien diagonal
diagonal = np.diagonal(np.abs(a))

# Untuk mencari pertambahan baris tanpa diagonal
off_diagonal = np.sum(np.abs(a), axis=1) - diagonal

if np.all(diagonal > off_diagonal):
  print('Matrix is diagonally dominant')
else:
  print('Not diagonally dominant')
```

```python
x1 = 0
x2 = 0
x3 = 0
epsilon = 0.00001
converged = False

x_old = np.array([x1, x2, x3])

print('Hasil Iterasi')
print('k, x1, x2, x3')
for k in range(1, 50):
  x1 = (14-3*x2+3*x3)/(8)
  x2 = (5+2*x1-5*x3)/(-8)
  x3 = (-8-3*x1-5*x2)/(10)
  x = np.array([x1, x2, x3])

  # Cek apabila lebih kecil dari threshold
  dx = np.sqrt(np.dot(x-x_old, x-x_old))

  print("%d, %.4f, %.4f, %.4f"%(k, x1, x2, x3))
  if dx < epsilon:
    converged = True
    print('Converged!')
    break

  # Tampilkan nilai x terakhir ke nilai x lama
  x_old = x

if not converged:
  print('Not converged, increase the # of iterations')
```
